```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = (10, 6), facecolor = (0.3 ,1 ,1), edgecolor = 'green', linewidth = 5) # 크기 설정
x = np.arange(0, 10, 0.1)
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.title('Example of sinewave')      # 제목
plt.grid()                            # 격자표시
plt.legend()                          # 범례표시
plt.xlabel('Time')                    # x축 라벨
plt.ylabel('Amplitude')               # y축 라벨
# frameon = False 버그인듯
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)   # subplots = subplot + figure 이런방식이고 여러개 생성
fig.tight_layout()                             # 눈금을 자동조절해서 보기좋게 해줌
x = np.arange(10) 
ax[0][0].plot(x, x) 
ax[0][1].scatter(x, x) 
ax[1][0].plot(x, x ** 2) 
ax[1][1].scatter(x, x ** 2)
```
