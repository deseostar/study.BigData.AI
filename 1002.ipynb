```python
def getRegion(gu):
    도심권 = '종로구 중구 용산구'.split()
    서북권 = '은평구 서대문구 마포구'.split()
    서남권 = '강서구 양천구 영등포구 구로구 동작구 관악구 금천구'.split()
    동남권 = '서초구 강남구 송파구 강동구'.split()
    동북권 = '도봉구 노원구 중랑구 광진구'.split() +\
             '성동구 동대문구 성북구 강북구'.split()
    if gu in 도심권: return '도심권'
    if gu in 서북권: return '서북권'
    if gu in 서남권: return '서남권'
    if gu in 동남권: return '동남권'
    if gu in 동북권: return '동북권'
 
pop_in_Seoul['권역'] = pop_in_Seoul.index.map(getRegion)    
pop_in_Seoul.tail()
```

```python
import googlemaps
import googlemaps_key
gmaps = googlemaps.Client(key = googlemaps_key.gmaps_key)
```

```python
gmaps.geocode('서울시청', language = 'ko')
```

```python
def getAddr(name):
    geo = gmaps.geocode(name, language = 'ko')
    return geo[0].get('formatted_address')
def getLoc(name):
    geo = gmaps.geocode(name, language = 'ko')
    tmp = geo[0].get('geometry').get('location')
    return tmp.get('lat'), tmp.get('lng')
```
