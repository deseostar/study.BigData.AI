```python
file2 = 'C:/Users/deseo/AI ex Folder/data/pop_in_Seoul.txt'
pop_in_Seoul = pd.read_csv(file2, encoding = 'utf-8', sep = '\t', header = 2, usecols = [1, 3, 6, 9, 13])
pop_in_Seoul.columns = ['구별', '인구수', '한국인', '외국인', '고령자']
pop_in_Seoul
```

```python
pop_in_Seoul.columns[1:]
```

```python
for c in pop_in_Seoul.columns[1:]:
    pop_in_Seoul[c] = pop_in_Seoul[c].str.replace(',', '').astype('int')
pop_in_Seoul.tail()
```

```python
pop_in_Seoul.set_index('구별', inplace = True)
pop_in_Seoul.tail()
```

```python
pop_in_Seoul.drop('합계', inplace = True)
pop_in_Seoul.tail()
```
